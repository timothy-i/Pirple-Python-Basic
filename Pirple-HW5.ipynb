```python
for number in range(1,101):
    if number % 15 == 0:
        print('FizzBuzz','(',number,')')
        continue
    if number % 3 == 0:
        print('Fizz','(',number,')')
        continue
    if number % 5 == 0:
        print('Buzz','(',number,')')
        continue
    if number % number == 0 and number % 2 == 1:
        print('prime','(',number,')')
        continue
    print(number)
```
