```python
fileNameList = []

while True:
    fileName = input('Please enter your file name:\n')
    if fileName == 'stop':
        break
        
    elif fileName not in fileNameList:
        fileNameList.append(fileName)
        with open(fileName + '.txt', 'w') as noteFile:
            note = input('Enter your note:\n')
            noteFile.write(note + '\n')
            print('Note saved!')
        
    elif fileName in fileNameList:
        option = input('Your note is already exist,\ndo you want to read the file, overwrite the file, append the file, or replace a single line?\n')
        if option == 'read':
            with open(fileName + '.txt', 'r') as noteFile:
                for line in noteFile:
                    print(line)
        elif option == 'overwrite':
            with open(fileName + '.txt', 'w') as noteFile:
                note = input('Enter your note:\n')
                noteFile.write(note)
                print('Note saved!')
        elif option == 'append':
            with open(fileName + '.txt', 'a') as noteFile:
                note = input('Enter your note:\n')
                noteFile.write(note + '\n')
                print('Note updated!')
        elif option == 'replace':
            lineChange = int(input('Which line would you like to replace? '))
            note = input('Enter your note:\n')
            tempNote = []
            with open(fileName + '.txt',"r") as noteFile:
                for line in noteFile:
                    tempNote.append(line)
            tempNote[lineChange-1] = note + "\n"
            with open(fileName + '.txt', "w") as noteFile:
                for i in tempNote:
                    noteFile.write(i)
                print('Note updated!')
        else:
            print('Please enter read, overwrite, append, or replace')
    
```
