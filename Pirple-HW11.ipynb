```python
def check1(a,b,c):
    if a == b:
        print(True)
    elif a == c:
        print(True)
    elif b == c:
        print(True)
    else:
        print(False)
        
def check(a = None, b = None, c = None):
    try:
        rslt = False
        a = int(a)
        b = int(b)
        c = int(c)
        print('accepted')
        if a == b or b == c or c == a:
            rslt = True
    except ValueError:
        print('ValueError: Plase input numbers only')
    except TypeError:
        print('TypeError: Please provide 3 inputs & numbers only')
        raise
    except Exception as e:
        print(e)
    finally:
        print('Input: ', a, b, c)
        print('Results: ', rslt)
        return rslt


#check one string
check('a',1,2)
print('\n')

#check string only
check('a','b','c')
print('\n')

#check 1 input only
check(1)
print('\n')

#check str, float, int input
check('a',1.2,3)
print('\n')

#check Boolean input
check(True,False,False)
print('\n')

#check 1 bool input only
check(False)
print('\n')

#check list input
check([1],[2],[3])
print('\n')

#check dict input
check({1:'one'},{2:'two'},{3:'three'})
print('\n')

#check no input
check()
print('\n')

#check empty str
check(' ',' ',' ')
print('\n')
```
