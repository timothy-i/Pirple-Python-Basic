```python
import statistics as st
from sty import fg, bg, ef, rs

'''Air Quality Index Statistics'''

##AQI last 30 days
#aqiData=[92,90,81,74,78,75,63,71,104,86,75,75,80,90,107,115,109,77,68,71,90,103,99,96,114,97,64,81,85,117]
#title = ef.bold + 'Surabaya Air Quality Index Statistics in last 30 days' + rs.bold_dim

#Input AQI data
aqiData = []
revise = 'yes'
city = ''

while revise == 'yes':
    city = input('Please enter the city: ')
    for i in range(30):
        inputAqi = int(input('Please enter 30 days samples of the Air Quality Index: '))
        aqiData.append(inputAqi)
    revise = input('Do you want to revise the whole data? \n')

title = ef.bold + city + ' Air Quality Index Statistics in last 30 days' + rs.bold_dim

#Calculate the mean
mean = st.mean(aqiData)

#Calculate the geometric mean
gMean = st.geometric_mean(aqiData)

#Calculate the median
med = st.median(aqiData)

#Calculate the mode
mode = st.mode(aqiData)

#Calculate the standart deviation
std = st.stdev(aqiData)

#Calculate the variance
var = st.variance(aqiData)

#Find the best and worst day of AQI
best = min(aqiData)
indexB = aqiData.index(best) + 1

worst = max(aqiData)
indexW = aqiData.index(worst) + 1

#Categorization
def aqiColor():
    aqiColorLevels = []
    for i in range(len(aqiData)):
        if aqiData[i] <= 50:
            colorLevels = bg.green + str(aqiData[i]) + rs.bg
            aqiColorLevels.append(colorLevels)
    
        if aqiData[i] <= 100 and aqiData[i] >= 51:
            colorLevels = bg.yellow + str(aqiData[i]) + rs.bg
            aqiColorLevels.append(colorLevels)

        if aqiData[i] <= 150 and aqiData[i] >= 101:
            colorLevels = bg(208) + str(aqiData[i]) + rs.bg
            aqiColorLevels.append(colorLevels)
    
        if aqiData[i] <= 200 and aqiData[i] >= 151:
            colorLevels = bg.red + str(aqiData[i]) + rs.bg
            aqiColorLevels.append(colorLevels)
    
        if aqiData[i] <= 300 and aqiData[i] >= 201:
            colorLevels = bg.da_magenta + str(aqiData[i]) + rs.bg
            aqiColorLevels.append(colorLevels)
    
        if aqiData[i] >= 301:
            colorLevels = bg.da_red + str(aqiData[i]) + rs.bg
            aqiColorLevels.append(colorLevels)
    
    print(', '.join(aqiColorLevels))

#main
print('')
print(title)
aqiColor()
print('')
print('Mean of AQI:', mean)
print('Geometric mean of AQI:', gMean)
print('Median of AQI:', med)
print('Mode of AQI:', mode)
print('Standart Deviation of AQI:', std)
print('Variance of AQI:', var)
print('Best day for last 30 days:', 'Day', indexB)
print('Worst day for last 30 days:', 'Day', indexW)
```
